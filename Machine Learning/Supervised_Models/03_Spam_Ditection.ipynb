```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('./sms_spam_data/spam.txt', delimiter="\t", names=['lable', 'sms'])
```

```python
x = df['sms']
y = df['lable']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, )
print(x_train.head(2),end="\n\n\n")
print(y_train.head(2))
```

```python
matrixer = CountVectorizer()
x_train = matrixer.fit_transform(x_train)
```

```python
x_train.todense()
```

```python
pd.DataFrame(
    x_train.todense(),
    columns=matrixer.get_feature_names_out()
).head(3)
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
classifier = BernoulliNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(matrixer.fit_transform(x_test))
```
